```python
import statistics
import pandas as pd
import csv
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
math_list = df['math score'].tolist()
reading_list = df['reading score'].tolist()
writing_list = df['writing score'].tolist()

```

```python
from os import read
math_mean = statistics.mean(math_list)
reading_mean = statistics.mean(reading_list)
writing_mean = statistics.mean(writing_list)

math_med = statistics.median(math_list)
reading_med = statistics.median(reading_list)
writing_med = statistics.median(writing_list)

math_mode = statistics.mode(math_list)
reading_mode = statistics.mode(reading_list)
writing_mode = statistics.mode(writing_list)

print("mean, median, and mode of math scores are {}, {}, {} resectively".format(math_mean, math_med, math_mode))
print("mean, median, and mode of reading scores are {}, {}, {} resectively".format(reading_mean, reading_med, reading_mode))
print("mean, median, and mode of reading scores are {}, {}, {} resectively".format(writing_mean, writing_med, writing_mode))


```

```python
math_st = statistics.stdev(math_list)
reading_st = statistics.stdev(reading_list)
writing_st = statistics.stdev(writing_list)

print("The standard deviation of math scores are {} respectively".format(math_st))
print("The standard deviation of reading scores are {} respectively".format(reading_st))
print("The standard deviation of writing scores are {} respectively".format(writing_st))
```

```python
math_fsd_start, math_fsd_end = math_mean - math_st, math_mean + math_st
reading_fsd_start, reading_fsd_end = reading_mean - reading_st, reading_mean + reading_st
writing_fsd_start, writing_fsd_end = writing_mean - writing_st, writing_mean + writing_st

math_list_of_data_within_1_std_deviation = [result for result in math_list if result > math_fsd_start and result < math_fsd_end]
reading_list_of_data_within_1_std_deviation = [result for result in reading_list if result > writing_fsd_start and result < reading_fsd_end]
writing_list_of_data_within_1_std_deviation = [result for result in writing_list if result > writing_fsd_start and result < writing_fsd_end]

print("{}% of data for math lies within 1 standard deviation".format(len(math_list_of_data_within_1_std_deviation)*100.0/len(math_list)))
print("{}% of data for reading lies within 1 standard deviation".format(len(reading_list_of_data_within_1_std_deviation)*100.0/len(reading_list)))
print("{}% of data for writing lies within 1 standard deviation".format(len(writing_list_of_data_within_1_std_deviation)*100.0/len(writing_list)))
```

```python
math_ssd_start, math_ssd_end = math_mean - (2*math_st), math_mean + (2*math_st)
reading_ssd_start, reading_ssd_end = reading_mean - (2*reading_st), reading_mean + (2*reading_st)
writing_ssd_start, writing_ssd_end = writing_mean - (2*writing_st), writing_mean + (2*writing_st)

math_list_of_data_within_2_std_deviations = [result for result in math_list if result > math_ssd_start and result < math_ssd_end]
reading_list_of_data_within_2_std_deviations = [result for result in reading_list if result > writing_ssd_start and result < reading_ssd_end]
writing_list_of_data_within_2_std_deviations = [result for result in writing_list if result > writing_ssd_start and result < writing_ssd_end]

print("{}% of data for math lies within 2 standard deviations".format(len(math_list_of_data_within_2_std_deviations)*100.0/len(math_list)))
print("{}% of data for reading lies within 2 standard deviations".format(len(reading_list_of_data_within_2_std_deviations)*100.0/len(reading_list)))
print("{}% of data for writing lies within 2 standard deviations".format(len(writing_list_of_data_within_2_std_deviations)*100.0/len(writing_list)))
```

```python
math_tsd_start, math_ssd_end = math_mean - (3*math_st), math_mean + (3*math_st)
reading_tsd_start, reading_tsd_end = reading_mean - (3*reading_st), reading_mean + (3*reading_st)
writing_tsd_start, writing_tsd_end = writing_mean - (3*writing_st), writing_mean + (3*writing_st)

math_list_of_data_within_3_std_deviations = [result for result in math_list if result > math_tsd_start and result < math_ssd_end]
reading_list_of_data_within_3_std_deviations = [result for result in reading_list if result > writing_tsd_start and result < reading_tsd_end]
writing_list_of_data_within_3_std_deviations = [result for result in writing_list if result > writing_tsd_start and result < writing_tsd_end]

print("{}% of data for math lies within 3 standard deviations".format(len(math_list_of_data_within_3_std_deviations)*100.0/len(math_list)))
print("{}% of data for reading lies within 3 standard deviations".format(len(reading_list_of_data_within_3_std_deviations)*100.0/len(reading_list)))
print("{}% of data for writing lies within 3 standard deviations".format(len(writing_list_of_data_within_3_std_deviations)*100.0/len(writing_list)))
```
